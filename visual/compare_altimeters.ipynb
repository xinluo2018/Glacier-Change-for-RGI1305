```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
```

```python
path_isat1 = 'data/icesat-1/tiles-dif-srtm/tile_30_95.h5'
path_isat2 = 'data/icesat-2/tiles-dif-srtm/tile_30_95.h5'
```

```python
dif_isat1, dif_isat2 = {}, {}
with h5py.File(path_isat1, 'r') as f_read:
    for key in f_read.keys():
        dif_isat1[key] = f_read[key][:]
with h5py.File(path_isat2, 'r') as f_read:
    for key in f_read.keys():
        dif_isat2[key] = f_read[key][:]
dif_isat1.keys()
```

```python
ids_isat1_stable = np.where((dif_isat1['type_fp']==1))[0]
ids_isat2_stable = np.where(dif_isat2['type_fp']==1)[0]
dif_isat1_stable = dif_isat1['h_dif'][ids_isat1_stable]
dif_isat2_stable = dif_isat2['h_dif'][ids_isat2_stable]
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
counts_bin_isat1 = plt.hist(dif_isat1_stable, bins=50, color='blue')
plt.xlim(left=-50, right=50)
plt.subplot(1,2,2)
counts_bin_isat2 = plt.hist(dif_isat2_stable, bins=100, color='green')
plt.xlim(left=-50, right=50)
```

```python
print('icesat1 mean and std:', np.mean(dif_isat1_stable), np.std(dif_isat1_stable))
print('icesat2 mean and std:', np.mean(dif_isat2_stable), np.std(dif_isat2_stable))

```
