```python
path_dems_mosaic_2000 = 'data/aster_data/wkunlun-dems/dems_mosaic_2000.tif'
path_dems_mosaic_2001 = 'data/aster_data/wkunlun-dems/dems_mosaic_2001.tif'
path_dems_mosaic_2005 = 'data/aster_data/wkunlun-dems/dems_mosaic_2005.tif'
path_dems_mosaic_2019 = 'data/aster_data/wkunlun-dems/dems_mosaic_2019.tif'
path_dems_mosaic_2020 = 'data/aster_data/wkunlun-dems/dems_mosaic_2020.tif'
path_dems_mosaic_2021 = 'data/aster_data/wkunlun-dems/dems_mosaic_2021.tif'
path_imgs = [path_dems_mosaic_2000, path_dems_mosaic_2001, path_dems_mosaic_2005,
            path_dems_mosaic_2019, path_dems_mosaic_2020, path_dems_mosaic_2021]
```
