```python
import os
os.chdir('..')
import config
import matplotlib.pyplot as plt
import numpy as np
from utils.geotif_io import readTiff
from utils.geotif_io import readTiff
from utils.imgShow import imgShow
from utils.lay_stack import lay_stack

```

```python
path_dems_mosaic_2019 = 'data/aster_data/wkunlun-dems/dems_mosaic_2019.tif'
path_dems_mosaic_2020 = 'data/aster_data/wkunlun-dems/dems_mosaic_2020.tif'
path_dems_mosaic_2021 = 'data/aster_data/wkunlun-dems/dems_mosaic_2021.tif'
dems_mosaic_2019, dem_mosaic_2019_info = readTiff(path_dems_mosaic_2019)
dems_mosaic_2019[dems_mosaic_2019==0] = np.nan
dems_mosaic_2020, dem_mosaic_2020_info = readTiff(path_dems_mosaic_2020)
dems_mosaic_2020[dems_mosaic_2020==0] = np.nan
dems_mosaic_2021, dem_mosaic_2021_info = readTiff(path_dems_mosaic_2021)
dems_mosaic_2021[dems_mosaic_2021==0] = np.nan
```

```python
plt.figure(figsize=(15,10))
plt.subplot(1,3,1)
plt.imshow(dems_mosaic_2019, vmin=3000, vmax=8000, cmap='terrain')
plt.title('dem_2019')
plt.subplot(1,3,2)
plt.imshow(dems_mosaic_2020, vmin=3000, vmax=8000, cmap='terrain')
plt.title('dem_2020')
plt.subplot(1,3,3)
plt.imshow(dems_mosaic_2021, vmin=3000, vmax=8000, cmap='terrain')
plt.title('dem_2021')
```
